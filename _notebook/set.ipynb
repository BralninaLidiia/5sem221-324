# Множества

```
HashSet<>//неупорядоченное множество
SortedSet<>//упорядоченное множество
```

```
HashSet<int> numbers=new();
numbers.Add(3);
numbers.Add(2);
numbers.Add(1);
numbers.Add(3);
numbers.Add(1);
numbers.Add(3);
numbers
```

```
numbers.Contains(1);
```

```
HashSet<int> a = new(){1, 3, 2, 4};
HashSet<int> b = new(){5, 7, 2, 4};
a.IntersectWith(b);
a
```

```
HashSet<int> a = new(){1, 3, 2, 4};
HashSet<int> b = new(){5, 7, 2, 4};
a.UnionWith(b);
a
```

```
HashSet<int> a = new(){1, 3, 2, 4};
HashSet<int> b = new(){5, 7, 2, 4};
a.ExceptWith(b);
a
```

```
SortedSet<int> numbers=new();
numbers.Add(3);
numbers.Add(2);
numbers.Add(1);
numbers.Add(3);
numbers.Add(1);
numbers.Add(3);
numbers
```
